```python
import os
from google import genai

client = genai.Client()

model_name = "gemini-2.5-flash"

prompt = "Explain the concept of quantum entanglement in simple terms."

response = client.models.generate_content(
    model=model_name,
    contents=prompt,
)

print(response.text)
```
